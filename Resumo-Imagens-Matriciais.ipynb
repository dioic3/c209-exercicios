```python
# Exercicio 1 - metodo grayscale e imagem resultante no disco
# primeiro - abra a imagem
# segundo - depois aplique o metodo
# exiba a imagem 
# depois salve a imagem
img = Image.open('dioice.jpg')
img_np = np.array(img)
def grayscale(img_np):
    (l, c, p) = img_np.shape
    img_avg = np.zeros(shape=(l, c), dtype=np.uint8)
    for i in range(l):
        for j in range(c):
            r = float(img_np[i, j, 0])
            g = float(img_np[i, j, 1])
            b = float(img_np[i, j, 2])

            img_avg[i, j] = (r + g + b) / 3
        
    return img_avg

img_avg = grayscale(img_np)
plt.imshow(img_avg,cmap = 'gray')
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

img = Image.open('link.jpg') #abra a imagem
img1 = np.array(img) # convertar em vetor

def binaria(img_np):
    (l, c, p) = img_np.shape
    img_avg = np.zeros(shape=(l, c), dtype=np.uint8)
    for i in range(l):
        for j in range(c):
            r = float(img_np[i, j, 0])
            g = float(img_np[i, j, 1])
            b = float(img_np[i, j, 2])
            img_avg[i, j] = (max(r, g, b) + min(r, g, b)) / 3
            
    img_bin = np.zeros(shape=(l, c), dtype=np.uint8)
    tresh = 10
    for i in range(l):
        for j in range(c):
            if img_avg[i, j] > tresh:
                img_bin[i, j] = 255 
    return img_bin
    

img_bin = binaria(img1)
plt.imshow(img_bin, cmap = 'gray')

img_pil = Image.fromarray(img_bin)
img_pil.save('link1.png')
```
