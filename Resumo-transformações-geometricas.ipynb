```python
# Exercicio 1
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

mario = np.array(Image.open('mario.png'))[:, :, :3]
def mirror(mario,reserve_x,reserve_y):
    (l, c, p) = mario.shape
    mario_refl = np.zeros(shape=mario.shape, dtype=np.uint8)
    if reserve_x == True:
        for i in range(l):
            for j in range(c):
                new_x = -i
                new_y = j
                mario_refl[new_y, new_x] = mario[i, j]
        return mario_refl
    elif reserve_y == True:
          for i in range(l):
            for j in range(c):
                new_x = j
                new_y = -i
                mario_refl[new_y, new_x] = mario[i, j]
            return mario_refl
    elif reserve_x == True and reserve_y == True:
        for i in range(l):
            for j in range(c):
                new_x = -j
                new_y = -i
                mario_refl[new_y, new_x] = mario[i, j]
            return mario_refl
        

img = mirror(mario,True,False)
plt.imshow(img)
```

```python
# Exercicio 2
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

mario = np.array(Image.open('mario.png'))[:, :, :3]
bx, by = 1, 2
def cisalhamento(mario):
    (l, c, p) = mario.shape
    mario_refl = np.zeros(shape=((int(l* (1 + bx))),int(c *(1+by)),p),dtype = np.uint8)
    for i in range(l):
        for j in range(c):
            new_x = int(j + i * by)
            new_y = int(i + j * bx)
            mario_refl[new_y, new_x] = mario[i, j]
    return mario_refl

img = cisalhamento(mario)
plt.imshow(img)
```
